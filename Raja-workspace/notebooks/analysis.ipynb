```python
# ================== 1. SETUP ==================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd

# For interactive maps
import folium
from folium.plugins import HeatMap

# Paths
f_properties = "/Users/rajaa/Desktop/Applied Data Science MAST30034/project-2-group-real-estate-industry-project-7-2025/Raja-workspace/notebooks/data/raw/tidy_monthly/enriched/all_properties_tidy_enriched.csv"
f_crime      = "/Users/rajaa/Desktop/Applied Data Science MAST30034/project-2-group-real-estate-industry-project-7-2025/Raja-workspace/notebooks/data/cleaned_tasneem_ariq/3. crime/crime dataset.csv"
f_demo       = "/Users/rajaa/Desktop/Applied Data Science MAST30034/project-2-group-real-estate-industry-project-7-2025/Raja-workspace/notebooks/data/cleaned_tasneem_ariq/4. demographic/demographic_dataset.csv"

# ================== 2. LOAD DATA ==================
df_properties = pd.read_csv(f_properties)
df_crime      = pd.read_csv(f_crime)
df_demo       = pd.read_csv(f_demo)

print("Properties:", df_properties.shape)
print("Crime:", df_crime.shape)
print("Demographics:", df_demo.shape)

# ================== 3. QUICK LOOK ==================
display(df_properties.head())
display(df_crime.head())
display(df_demo.head())

# ================== 4. CLEANING ==================
# Ensure consistent suburb naming
df_properties["Suburb"] = df_properties["Suburb"].str.strip().str.lower()
df_crime["Cluster"]     = df_crime["Cluster"].str.strip().str.lower()
df_demo["Target_Suburb"]= df_demo["Target_Suburb"].str.strip().str.lower()

# ================== 5. SUMMARY STATS ==================
print("\n--- Properties ---")
print(df_properties.describe(include="all"))

print("\n--- Crime ---")
print(df_crime.describe(include="all"))

print("\n--- Demographics ---")
print(df_demo.describe(include="all"))

# ================== 6. TIME SERIES (Property Prices) ==================
df_properties["date"] = pd.to_datetime(df_properties["date"])
trend = (df_properties
         .groupby(["date","property_type"])["Median"]
         .mean()
         .reset_index())

plt.figure(figsize=(12,6))
sns.lineplot(data=trend, x="date", y="Median", hue="property_type")
plt.title("Median Property Prices Over Time by Property Type")
plt.show()
```

```python

# ================== 7. CORRELATIONS ==================
# Merge datasets on suburb (approximate)
merged = df_properties.merge(
    df_demo, left_on="Suburb", right_on="Target_Suburb", how="left"
)
merged = merged.merge(
    df_crime, left_on="Suburb", right_on="Cluster", how="left"
)

# Select numerical columns only
num_cols = merged.select_dtypes(include=[np.number])
corr = num_cols.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap (Properties + Crime + Demographics)")
plt.show()
```

```python

# ================== 8. PROPERTY VS DEMOGRAPHICS ==================
plt.figure(figsize=(8,6))
sns.scatterplot(data=merged, x="average_income", y="Median", alpha=0.6)
plt.title("Property Median Price vs Average Income")
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(data=merged, x="tenant_rate", y="Median", alpha=0.6)
plt.title("Property Median Price vs Tenant Rate")
plt.show()
```

```python

# ================== 9. CRIME VS PROPERTIES ==================
plt.figure(figsize=(8,6))
sns.scatterplot(data=merged, x="crime_per_1000", y="Median", alpha=0.6)
plt.title("Property Median Price vs Crime Rate per 1000")
plt.show()
```

```python

# ================== 10. GEO VISUALISATION ==================
# Crime heatmap
crime_map = folium.Map(location=[-37.81, 144.96], zoom_start=9)  # Center on Melbourne
heat_data = merged[["Lat_x","Lng_x","crime_per_1000"]].dropna().values.tolist()
HeatMap(heat_data, radius=10).add_to(crime_map)
crime_map.save("crime_heatmap.html")

# Property median price map
price_map = folium.Map(location=[-37.81, 144.96], zoom_start=9)
for _, row in merged.dropna(subset=["Lat_x","Lng_x","Median"]).iterrows():
    folium.CircleMarker(
        location=[row["Lat_x"], row["Lng_x"]],
        radius=5,
        popup=f"{row['Suburb']}: {row['Median']}",
        color="blue",
        fill=True,
        fill_opacity=0.6
    ).add_to(price_map)
price_map.save("property_prices_map.html")

print("EDA complete. Check generated plots + 'crime_heatmap.html' and 'property_prices_map.html'.")
```

```python
# %% --- Trim datasets to 2025 onwards ----------------------------------------
import pandas as pd
from pathlib import Path

# Input and output dirs
WITH_PT_DIR = Path(
    "/Users/rajaa/Desktop/Applied Data Science MAST30034/"
    "project-2-group-real-estate-industry-project-7-2025/"
    "Raja-workspace/notebooks/data/raw/tidy_monthly/enriched/with_pt"
)
TRIMMED_DIR = WITH_PT_DIR / "trimmed_2025onward"
TRIMMED_DIR.mkdir(parents=True, exist_ok=True)

# Process all CSVs in with_pt/
csvs = sorted([p for p in WITH_PT_DIR.glob("*.csv") if not p.name.startswith("~$")])
print(f"[info] found {len(csvs)} CSV(s) in {WITH_PT_DIR}")

for f in csvs:
    try:
        df = pd.read_csv(f, parse_dates=["date"], dayfirst=False, infer_datetime_format=True)
        if "date" not in df.columns:
            print(f" - {f.name}: no 'date' column, skipping")
            continue

        # Keep only 2025+
        df_2025 = df[df["date"] >= pd.Timestamp("2025-01-01")].copy()

        if df_2025.empty:
            print(f" - {f.name}: all rows pre-2025, skipping output")
            continue

        out_path = TRIMMED_DIR / f.name
        df_2025.to_csv(out_path, index=False)
        print(f" âœ“ {f.name}: kept {len(df_2025)} rows â†’ {out_path.name}")

    except Exception as e:
        print(f" ! {f.name}: FAILED â€” {e}")

print("\n[done] Trimmed datasets saved in:", TRIMMED_DIR)
```

```python
import pandas as pd
from pathlib import Path

# --- Paths ---
BASE_DIR = Path(
    "/Users/rajaa/Desktop/Applied Data Science MAST30034/"
    "project-2-group-real-estate-industry-project-7-2025/"
    "Raja-workspace/notebooks/data/raw/tidy_monthly/enriched/with_pt"
)
TRIMMED_DIR = BASE_DIR / "trimmed_2025onward"
OUT_DIR = BASE_DIR / "trimmed_2025onward_enriched"
OUT_DIR.mkdir(parents=True, exist_ok=True)

# --- Load crime + demographics ---
crime = pd.read_csv(
    "/Users/rajaa/Desktop/Applied Data Science MAST30034/project-2-group-real-estate-industry-project-7-2025/"
    "Raja-workspace/notebooks/data/cleaned_tasneem_ariq/3. crime/crime dataset.csv"
)
demo = pd.read_csv(
    "/Users/rajaa/Desktop/Applied Data Science MAST30034/project-2-group-real-estate-industry-project-7-2025/"
    "Raja-workspace/notebooks/data/cleaned_tasneem_ariq/4. demographic/demographic_dataset.csv"
)

# --- Standardize join keys for crime & demo ---
for df in [crime, demo]:
    if "SA2_NAME21" in df.columns:
        df["SA2_NAME21"] = df["SA2_NAME21"].astype(str).str.upper().str.strip()
    if "SA2_CODE21" in df.columns:
        df["SA2_CODE21"] = (
            pd.to_numeric(df["SA2_CODE21"], errors="coerce")
            .fillna(-1).astype(int).astype(str)
        )

# --- Process rental datasets ---
for f in sorted(TRIMMED_DIR.glob("*.csv")):
    try:
        df = pd.read_csv(f, parse_dates=["date"])

        # Fix SA2_CODE21_x â†’ SA2_CODE21
        if "SA2_CODE21_x" in df.columns and "SA2_CODE21" not in df.columns:
            df = df.rename(columns={"SA2_CODE21_x": "SA2_CODE21"})

        # Case 1: join by code
        if "SA2_CODE21" in df.columns:
            df["SA2_CODE21"] = (
                pd.to_numeric(df["SA2_CODE21"], errors="coerce")
                .fillna(-1).astype(int).astype(str)
            )
            merged = (
                df.merge(crime, on="SA2_CODE21", how="left", suffixes=("", "_crime"))
                  .merge(demo, on="SA2_CODE21", how="left", suffixes=("", "_demo"))
            )

        # Case 2: join by name
        elif "SA2_NAME21" in df.columns:
            df["SA2_NAME21"] = df["SA2_NAME21"].astype(str).str.upper().str.strip()
            merged = (
                df.merge(crime, on="SA2_NAME21", how="left", suffixes=("", "_crime"))
                  .merge(demo, on="SA2_NAME21", how="left", suffixes=("", "_demo"))
            )

        # Case 3: no key â†’ skip
        else:
            print(f" ! {f.name}: skipped (no SA2_CODE21 or SA2_NAME21 column)")
            continue

        out_path = OUT_DIR / f.name
        merged.to_csv(out_path, index=False)
        print(f" âœ“ {f.name} enriched with crime+demo â†’ {out_path.name}")

    except Exception as e:
        print(f" ! {f.name}: FAILED â€” {e}")

print("\n[done] All enriched datasets saved in:", OUT_DIR)
```

```python
import pandas as pd
from pathlib import Path

import pandas as pd
from pathlib import Path

# --- Paths ---
BASE_DIR = Path(
    "/Users/rajaa/Desktop/Applied Data Science MAST30034/"
    "project-2-group-real-estate-industry-project-7-2025/"
    "Raja-workspace/notebooks/data/raw/tidy_monthly/enriched/with_pt"
)
CLEANED_DIR = Path(BASE_DIR) / "cleaned_with_pt"   # ensure it's a Path
OUT_DIR     = Path(BASE_DIR) / "trimmed_2025onward_enriched"
OUT_DIR.mkdir(parents=True, exist_ok=True)

# --- Load crime + demographics ---
crime = pd.read_csv(
    "/Users/rajaa/Desktop/Applied Data Science MAST30034/project-2-group-real-estate-industry-project-7-2025/"
    "Raja-workspace/notebooks/data/cleaned_tasneem_ariq/3. crime/crime dataset.csv"
)
demo  = pd.read_csv(
    "/Users/rajaa/Desktop/Applied Data Science MAST30034/project-2-group-real-estate-industry-project-7-2025/"
    "Raja-workspace/notebooks/data/cleaned_tasneem_ariq/4. demographic/demographic_dataset.csv"
)

# --- Standardise SA2 in crime + demo ---
for df_ext in [crime, demo]:
    # rename "SA2" â†’ "SA2_CODE21" if needed
    if "SA2" in df_ext.columns and "SA2_CODE21" not in df_ext.columns:
        df_ext.rename(columns={"SA2": "SA2_CODE21"}, inplace=True)

    if "SA2_NAME21" in df_ext.columns:
        df_ext["SA2_NAME21"] = df_ext["SA2_NAME21"].astype(str).str.upper().str.strip()

    if "SA2_CODE21" in df_ext.columns:
        df_ext["SA2_CODE21"] = (
            pd.to_numeric(df_ext["SA2_CODE21"], errors="coerce")
            .astype("Int64")
            .astype(str)
        )

# --- Helper: collapse duplicate columns (x/y) ---
def coalesce_cols(df, primary, secondary, new_name):
    if primary in df.columns and secondary in df.columns:
        df[new_name] = df[primary].where(
            df[primary].notna() & (df[primary].astype(str) != "nan"),
            df[secondary]
        )
        df.drop(columns=[primary, secondary], inplace=True)
    elif primary in df.columns:
        df.rename(columns={primary: new_name}, inplace=True)
    elif secondary in df.columns:
        df.rename(columns={secondary: new_name}, inplace=True)
    return df

# --- Process rental datasets from CLEANED_DIR ---
files = sorted(p for p in CLEANED_DIR.glob("*.csv") if not p.name.startswith("~$"))
print(f"[info] found {len(files)} CSV(s) in {CLEANED_DIR}")

for f in files:
    try:
        df = pd.read_csv(f, parse_dates=["date"])

        # Ensure codes are clean strings if present
        for c in ["SA2_CODE21", "LGA_CODE21"]:
            if c in df.columns:
                df[c] = pd.to_numeric(df[c], errors="coerce").astype("Int64").astype(str)

        # Coalesce leftover x/y columns
        for cpri, csec, cname in [
            ("SA2_CODE21_x","SA2_CODE21_y","SA2_CODE21"),
            ("LGA_CODE21_x","LGA_CODE21_y","LGA_CODE21"),
            ("SA2_NAME21_x","SA2_NAME21_y","SA2_NAME21"),
            ("LGA_NAME21_x","LGA_NAME21_y","LGA_NAME21"),
        ]:
            df = coalesce_cols(df, cpri, csec, cname)

        # Standardise names
        for c in ["SA2_NAME21", "LGA_NAME21"]:
            if c in df.columns:
                df[c] = df[c].astype(str).str.upper().str.strip()

        # --- Merge with crime + demographics ---
        if "SA2_CODE21" in df.columns:
            merged = (
                df.merge(crime, on="SA2_CODE21", how="left", suffixes=("", "_crime"))
                  .merge(demo,  on="SA2_CODE21", how="left", suffixes=("", "_demo"))
            )
            key_used = "SA2_CODE21"
        elif "SA2_NAME21" in df.columns:
            merged = (
                df.merge(crime, on="SA2_NAME21", how="left", suffixes=("", "_crime"))
                  .merge(demo,  on="SA2_NAME21", how="left", suffixes=("", "_demo"))
            )
            key_used = "SA2_NAME21"
        else:
            print(f" ! {f.name}: skipped (no SA2_CODE21 or SA2_NAME21 present)")
            continue

        # --- Diagnostics ---
        n_rows = len(df)
        matched_crime = merged.filter(regex="_crime$").notna().any(axis=1).sum()
        matched_demo  = merged.filter(regex="_demo$").notna().any(axis=1).sum()
        print(f" âœ“ {f.name}: key={key_used} | rows={n_rows} | crime match={matched_crime}/{n_rows} | demo match={matched_demo}/{n_rows}")

        # Save
        out_path = OUT_DIR / f.name
        merged.to_csv(out_path, index=False)

    except Exception as e:
        print(f" ! {f.name}: FAILED â€” {e}")

print("\n[done] Enriched datasets saved in:", OUT_DIR)
```
