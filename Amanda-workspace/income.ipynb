Importing libaries and income dataset per SA2 area code

```python
import pandas as pd
from pathlib import Path
DATA =  Path("..") / "datasets" / "VMFEAT"
income_df = pd.read_excel(DATA /"pop_income.xlsx", skiprows=6,sheet_name="Table 1.4",skipfooter=6) #Skipping header & footer and only taking from SA2 sheet 
```

Filtering to only victoria as the raw dataset has other states as well

```python
income_df
```

```python
# Find the row index where "Victoria" appears
vic_start = income_df[income_df.iloc[:,0] == "Victoria"].index[0]

# Subset everything after "Victoria"
vic_df = income_df.loc[vic_start:] 

# Drop rows after the next state header (e.g. "Queensland", "South Australia")
state_rows = vic_df[vic_df.iloc[:,0].isin(["Queensland","South Australia","Tasmania","Western Australia","Northern Territory","Australian Capital Territory"])]

if not state_rows.empty:
    vic_end = state_rows.index[0]
    vic_df = vic_df.loc[:vic_end-1]

# Reset index
vic_df = vic_df.reset_index(drop=True)
```

```python
vic_df
```

Renaming column and removing median age of earners column and earners (person) feature

```python
#Remove earners (person)
columns_to_remove = ["2017-18", "2018-19", "2019-20", "2020-21", "2021-22"]
vic_df = vic_df.drop(columns=columns_to_remove)
```

```python
#Remove meadian age earners
columns_to_remove = ["2017-18.1", "2018-19.1", "2019-20.1", "2020-21.1", "2021-22.1"]
vic_df = vic_df.drop(columns=columns_to_remove)
vic_df
```

```python
vic_df = vic_df.rename(columns=lambda x: x.replace(".2", ".sum") if x.endswith(".2") else x)
vic_df = vic_df.rename(columns=lambda x: x.replace(".3", ".med") if x.endswith(".3") else x)
vic_df = vic_df.rename(columns=lambda x: x.replace(".4", ".mean") if x.endswith(".4") else x)
vic_df
```

```python
#Checking for NULL values
vic_df.isnull().sum()

#No Null values, only 1 in the 'SA2 NAME' column 
#which belongs to Victoria (Make sense, do not have to handle this)
```

Convert income to numerical values

```python
cols_to_clean = vic_df.columns[2:]  # all columns after SA2 and SA2 NAME

for col in cols_to_clean:  # or sum_cols
    vic_df[col] = pd.to_numeric(
        vic_df[col].astype(str).str.replace(",", ""), 
        errors='coerce'  # convert bad values to NaN
    ).fillna(0) 

vic_df
```

```python
med_cols = [col for col in vic_df.columns if col.endswith(".med")]
mean_cols = [col for col in vic_df.columns if col.endswith(".mean")]

df_med_growth = vic_df[med_cols].pct_change(axis=1) * 100
df_mean_growth = vic_df[mean_cols].pct_change(axis=1) * 100

n_years = 4  
def calculate_cagr(start, end, n_years, eps=1e-9):
    start_safe = start + eps
    return ((end / start_safe) ** (1/n_years) - 1) * 100


vic_df['med_CAGR'] = vic_df.apply(lambda x: calculate_cagr(x[med_cols[0]], x[med_cols[-1]], n_years), axis=1)
vic_df['mean_CAGR'] = vic_df.apply(lambda x: calculate_cagr(x[mean_cols[0]], x[mean_cols[-1]], n_years), axis=1)

print(vic_df[['SA2 NAME', 'med_CAGR', 'mean_CAGR']])
```

# Predict income for the near 2030

```python
#Using CAGR
latest_year_col = med_cols[-1]  
latest_year = 2022  # the numeric year corresponding to 2021-22
target_year = 2030
years_to_2030 = target_year - latest_year  # 8 years

# Convert CAGR from % to decimal
vic_df['Predicted_Income_2030_CAGR'] = vic_df[latest_year_col] * (1 + vic_df['med_CAGR']/100) ** years_to_2030

# Show result
print(vic_df[['SA2 NAME', latest_year_col, 'med_CAGR', 'Predicted_Income_2030_CAGR']])
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

years_numeric = np.array([2018, 2019, 2020, 2021, 2022]).reshape(-1, 1)

# Years to predict
future_years = np.arange(2023, 2031).reshape(-1, 1) 

predictions_all_years = []

for idx, row in vic_df.iterrows():
    y = row[med_cols].values.astype(float)
    
    # Fit linear regression
    model = LinearRegression()
    model.fit(years_numeric, y)
    
    # Predict for all future years
    pred_future = model.predict(future_years)
    
    # Save as a list 
    predictions_all_years.append(pred_future)

future_years_cols = [f'Predicted_Income_{year}' for year in range(2023, 2031)]

pred_df = pd.DataFrame(predictions_all_years, columns=future_years_cols)

# Combine with original DataFrame
vic_df = pd.concat([vic_df.reset_index(drop=True), pred_df], axis=1)
```
