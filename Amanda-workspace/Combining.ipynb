```python
import pandas as pd
properties_df = pd.read_csv('/home/eeamanda/project-2-group-real-estate-industry-project-7-2025/datasets/all_properties_tidy_enriched.csv')
income_df = pd.read_csv('/home/eeamanda/project-2-group-real-estate-industry-project-7-2025/Amanda-workspace/predicted_income_till_2030.csv')
```

```python
properties_df[properties_df["SA2_CODE21"] == 201011002]
```

```python
income_df["SA2"] = income_df["SA2"].astype(int)
income_df
```

```python
def clean_sa2(value):
    try:
        return int(value)  # keep as integer if numeric
    except (ValueError, TypeError):
        return pd.NA       # replace non-numeric with NA (blank)

# Apply cleaning
income_df["SA2 NAME"] = income_df.apply(
    lambda row: row["SA2 NAME"] if pd.notna(row["SA2 NAME"]) else (row["SA2"] if not str(row["SA2"]).isdigit() else pd.NA),
    axis=1
)

income_df["SA2"] = income_df["SA2"].apply(clean_sa2).astype("Int64")
```

```python
income_df
```

Merge with income dataset

```python
properties_features_df = properties_df.merge(
    income_df,
    left_on="SA2_CODE21",  
    right_on="SA2",       
    how="left"             
)
```
